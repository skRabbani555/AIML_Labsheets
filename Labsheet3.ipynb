```python
year = range(2000,2008)
apples = [0.895,0.91,0.919,0.926,0.929,0.934,0.936,0.937]
oranges = [0.962,0.941,0.930,0.918,0.908,0.907,0.904,0.901]
```

```python
plt.plot(year,apples)
plt.plot(year,oranges)
plt.xlabel('year')
plt.ylabel('yeild')
```

```python
plt.plot(year,apples)
plt.plot(year,oranges)
plt.xlabel('year')
plt.ylabel('yeild')
plt.title('Crop yeilds in Kanto')
plt.legend(['Apples','Oranges']);
```

```python
plt.plot(year,apples,marker='o')
plt.plot(year,oranges,marker='x')

plt.ylabel('yeild')
plt.title('Crop yeilds in Kanto')
plt.legend(['Apples','Oranges']);
```

```python
plt.plot(year,apples,marker='s' ,c='b',ls='-' , lw=2, ms=8, mew=2, mec='navy')
plt.plot(year,oranges,marker='o',c='r',ls='--',lw=3,ms=10,alpha=0.5)

plt.ylabel('yeild')
plt.xlabel('years')
plt.title('Crop yeilds in Kanto')
plt.legend(['Apples','Oranges']);
```

```python
plt.plot(year,apples,'s-b')
plt.plot(year,oranges,'o--r')

plt.ylabel('yeild')
plt.xlabel('years')
plt.title('Crop yeilds in Kanto')
plt.legend(['Apples','Oranges']);
```

```python

plt.plot(year,oranges,'or')
plt.title('Yeild of Oranges')
plt.xlabel('year')
plt.ylabel('yeild')
```

```python
plt.figure(figsize=(4,4))
plt.plot(year,oranges,'or')
plt.title('Yeild of Oranges')
plt.xlabel('year')
plt.ylabel('yeild');
```

```python
x = [3,1,3,12,2,4,4]
y = [3,2,1,4,5,6,7]
plt.bar(x,y)
plt.title("Bar Chart")
plt.legend(["bar"])
plt.show()
```

```python
x = [3,1,3,12,2,4,4]
y = [3,2,1,4,5,6,7]
plt.scatter(x,y)
plt.legend("A")
plt.xlabel('x')
plt.ylabel('y')
plt.title("Scatter Plot")
plt.show()
```
